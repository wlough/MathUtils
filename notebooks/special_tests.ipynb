# Spherical harmonics point cloud fit test

```python
from pymathutils.pyutils import fib_sphere
from pymathutils.special import fit_real_sh_coefficients_to_points, compute_all_real_Ylm
from pymathutils import thetaphi_from_xyz
import numpy as np
import pyvista as pv

XYZ = fib_sphere(100)
# XYZ[:, 2] *= .6
# XYZ += 0.1
l_max = 10
reg_lambda = 1e-10
C = fit_real_sh_coefficients_to_points(XYZ, l_max, reg_lambda)
ThetaPhi = thetaphi_from_xyz(XYZ)
Y = compute_all_real_Ylm(l_max, ThetaPhi)
XYZ1 = np.einsum("ni,pn->pi", C, Y)

cloud = pv.PolyData(XYZ)
cloud["z"] = 0*XYZ[:, 2]

cloud1 = pv.PolyData(XYZ1)
cloud1["z"] = 0*XYZ1[:, 2]+1

# Plot
plotter = pv.Plotter()
plotter.add_mesh(
    cloud,
    render_points_as_spheres=True,
    point_size=5,
    scalars="z",      # optional
    cmap="viridis",   # optional
)
plotter.add_mesh(
    cloud1,
    render_points_as_spheres=True,
    point_size=5,
    scalars="z",      # optional
    cmap="viridis",   # optional
)
plotter.show()
```
